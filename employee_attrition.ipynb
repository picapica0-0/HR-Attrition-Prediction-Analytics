```python
import pandas as pd
import numpy as np

df = pd.read_csv('employee.csv')
df.head()
```

```python
# Xem tổng số dòng, số cột và kiểu dữ liệu của từng cột
df.info()
# Thống kê nhanh các cột số (Lương, Tuổi, Thâm niên...) xem có gì bất thường không
df.describe()
# Kiểm tra xem có bao nhiêu dòng bị trùng lặp (Duplicate)
print(f"Số dòng bị trùng: {df.duplicated().sum()}")
```

```python
# Đếm số lượng giá trị thiếu trong từng cột
print(df.isnull().sum())
```

```python
# Ví dụ kiểm tra xem có ai có số năm làm việc vô lý so với tuổi không
loi_logic = df[df['YearsAtCompany'] >= df['Age']]
print(f"Số dòng bị lỗi logic thâm niên: {len(loi_logic)}")
```

```python
# Kiểm tra các cột có giá trị độc nhất
print(df['EmployeeCount'].unique())
print(df['Over18'].unique())
print(df['StandardHours'].unique())
```

```python
# Xóa 3 cột vô nghĩa
df = df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'])
```

```python
# Kiểm tra số tuổi
print("Age range:", np.sort(df['Age'].unique()))
# Kiểm tra Khoảng cách từ nhà đến công ty
print("Distance from home:", np.sort(df['DistanceFromHome'].unique()))
```

```python
# 1. Gom nhóm Độ tuổi (Age)
def group_age(age):
    if age < 30: return 'Under 30'
    elif age <= 40: return '30 - 40'
    elif age <= 50: return '41 - 50'
    else: return 'Over 50'

df['Age_Group'] = df['Age'].apply(group_age)

# 2. Gom nhóm Khoảng cách từ nhà đến công ty (DistanceFromHome)
def group_distance(dist):
    if dist <= 5: return 'Near (0-5 km)'
    elif dist <= 15: return 'Medium (6-15 km)'
    else: return 'Far (>15 km)'

df['Distance_Group'] = df['DistanceFromHome'].apply(group_distance)
```

```python
# Chuyển đổi các cột Điểm số thành Nhãn chữ
satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
df['EnvironmentSatisfaction_Label'] = df['EnvironmentSatisfaction'].map(satisfaction_map)
df['JobInvolvement_Label'] = df['JobInvolvement'].map(satisfaction_map)
df['JobSatisfaction_label'] = df['JobSatisfaction'].map(satisfaction_map)
df['RelationshipSatisfaction_Label'] = df['RelationshipSatisfaction'].map(satisfaction_map)

education_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
df['Education_Label'] = df['EnvironmentSatisfaction'].map(education_map)

performance_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
df['PerformanceRating_Label'] = df['PerformanceRating'].map(performance_map)

work_life_map = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
df['WorkLifeBalance_Label'] = df['WorkLifeBalance'].map(work_life_map)
```

```python
# Xuất file đã làm sạch
df.to_csv('ibm_hr_cleaned_for_tableau.csv', index=False)
```
